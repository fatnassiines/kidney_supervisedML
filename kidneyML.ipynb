DATA SCALING & NORMALIZING

```python
from sklearn.preprocessing import StandardScaler, normalize
scaler = StandardScaler()
scaled_x = scaler.fit_transform(x)
# Normalizing the Data
normalized_x = normalize(scaled_x)

# Converting the numpy array into a pandas DataFrame
normalized_x = pd.DataFrame(normalized_x, columns= x.columns)
normalized_x
```

K-MEANS CLUSTERING

```python
from sklearn.cluster import KMeans
sse = {}
for k in range(1, 15):
    kmeans = KMeans(n_clusters=k, max_iter=1000, n_init = 10).fit(normalized_x)
    sse[k] = kmeans.inertia_ # Inertia: Sum of distances of samples to their closest cluster center
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
silhouette_scores = []

for n_cluster in range(2, 8):
    silhouette_scores.append(
        silhouette_score(normalized_x, KMeans(n_clusters = n_cluster, n_init = 10).fit_predict(normalized_x)))

# Plotting a bar graph to compare the results
k = [2, 3, 4, 5, 6,7]
plt.bar(k, silhouette_scores)
plt.xlabel('Number of clusters', fontsize = 10)
plt.ylabel('Silhouette Score', fontsize = 10)
plt.show()
```
